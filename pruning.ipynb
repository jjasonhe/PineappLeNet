```python
num_train, num_val, num_test = 70, 10, 20
train_data,val_data,test_data = du.create_dicts(num_train, num_val, num_test)

f = h5.File('datasets/train.h5', 'r')

for i,p in enumerate(train_data):
    imgs = f[train_data[i]]
    imgs = np.transpose(imgs, (0,1,3,2,4))
    imgs = np.absolute(imgs).astype(np.float32)
#     max_imgs = np.amax(imgs, axis=(2,3,4))
#     imgs = imgs/max_imgs[:,np.newaxis,np.newaxis,np.newaxis,:]
    avg_diffs = np.zeros((imgs.shape[0]))
    for t in range(18):
        plt.imshow(imgs[20,t,:,:,0], cmap='gray')
        plt.show()
    for z in range(imgs.shape[0]):
        timeseries = imgs[z, :, :, :, :]
        diffs = np.zeros((timeseries.shape[0]-1, timeseries.shape[1], timeseries.shape[2], timeseries.shape[3]))
        for t in range(timeseries.shape[0]-1):
            diffs[t, :, :, :] = np.absolute(timeseries[t+1, :, :, :] - timeseries[t, :, :, :])
        avg_diffs[z] = np.mean(diffs)
    plt.plot(np.linspace(0,f[train_data[i]].shape[0],f[train_data[i]].shape[0]), avg_diffs)
    plt.show()
```
