```python
import os
import imageio
import tensorflow as tf
```

```python
def save_images_from_event_0(fn, tag, output_dir='./'):
    assert(os.path.isdir(output_dir))

    image_str = tf.placeholder(tf.string)
    im_tf = tf.image.decode_image(image_str)

    sess = tf.InteractiveSession()
    with sess.as_default():
        count = 0
        name = 0
        for e in tf.train.summary_iterator(fn):
            for v in e.summary.value:
                if v.tag == tag:
                    im = im_tf.eval({image_str: v.image.encoded_image_string})
                    output_fn = os.path.realpath('{}/image_0{:04d}.png'.format(output_dir, count))
                    print("Saving '{}'".format(output_fn))
                    imageio.imwrite(output_fn, im)
                    count += 1
```

```python
def save_images_from_event_1(fn, tag, output_dir='./'):
    assert(os.path.isdir(output_dir))

    image_str = tf.placeholder(tf.string)
    im_tf = tf.image.decode_image(image_str)

    sess = tf.InteractiveSession()
    with sess.as_default():
        count = 0
        for e in tf.train.summary_iterator(fn):
            for v in e.summary.value:
                if v.tag == tag:
                    print(count)
                    if count > 194:
                        im = im_tf.eval({image_str: v.image.encoded_image_string})
                        output_fn = os.path.realpath('{}/image_1{:04d}.png'.format(output_dir, count))
                        print("Saving '{}'".format(output_fn))
                        imageio.imwrite(output_fn, im)
                    count += 1
```
